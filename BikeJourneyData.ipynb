# Seed shape data
#### For task 2, a new file has been loaded into a panda data frame. Clusters are then assigned to each of the seeds, in which the visualisation of the results can be seen below.
### Task 2, Option 1: Gaussian Mixture

```python
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
sns.set()
```

```python
seeds = pd.read_csv('seeds.csv')
```

```python
gmm = GaussianMixture(n_components=2)
```

```python
fit = gmm.fit(seeds[['area', 'groove length']])
predict = gmm.predict(seeds[['area', 'groove length']])
```

```python
plt.scatter(seeds['area'], seeds['groove length'], c = predict)
```

```python
x_values = []
model_aic = []
model_bic = []

for k in range (1, 10):
    mixturemodel = GaussianMixture(k)
    mixturemodel.fit(seeds[['area', 'groove length']])
    model_aic.append(mixturemodel.aic(seeds[['area', 'groove length']]))
    model_bic.append(mixturemodel.bic(seeds[['area', 'groove length']]))
    x_values.append(k)
```

```python
plt.plot(x_values, model_aic)
plt.xlabel('Number of clusters')
plt.ylabel('AIC value')
```

```python
plt.plot(x_values, model_bic)
plt.xlabel('Number of clusters')
plt.ylabel('BIC value')
```

- First, I loaded the seeds into a pandas data frame, then imported the gaussianmixture which is required for probability distribution and set the number of mixture components to 2. Variable ‘fit’ is then used to estimate the model parameters of the seeds. Similar to this, ‘predict’ is used to predict label for data. A Scatter graph is then used to showcase the results of the seeds, ‘c = predict’ is used to colormap the predicted seeds. A for loop is then executed for the range of 9 values. In this loop mixture models based on the seeds are created to provide us with the probabilities that the given data point belongs to each of the clusters. Gaussian mixtures were ideal for this scenario as they do not hard-assign data points to a cluster much like k-means does, therefore, avoiding any mis grouping. Then we fit and calculate the AIC and BIC for each number of clusters (each number of clusters is considered a different model). The values for both the AIC and BIC are plotted.

### Task 2, Option 2: KMeans Algorithm

```python
from sklearn.cluster import KMeans
plt.scatter(seeds['area'], seeds['groove length'])
plt.xlabel('area')
plt.ylabel('groove length')
plt.show()
```

```python
x = seeds.copy()
```

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
clusters = x.copy()
clusters['cluster_pred'] = kmeans.fit_predict(x)
```

```python
plt.scatter(clusters['area'], clusters['groove length'], c = clusters['cluster_pred'], cmap = 'rainbow')
plt.xlabel('area')
plt.ylabel('groove length')
plt.show()
```

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

```python
wcss = []
```

```python
for i in range (1, 30):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1, 30),wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans_new = KMeans(3)
kmeans.fit(x_scaled)
cluster_new = x.copy()
cluster_new['cluster_pred'] = kmeans_new.fit_predict(x_scaled)
cluster_new
```

```python
plt.scatter(cluster_new['area'], cluster_new['groove length'], cluster_new['perimeter'], cluster_new['width'], cmap = 'rainbow')
plt.xlabel('Number of Clusters')
plt.ylabel('Loyalty')
plt.show()
```

- In this method, I used the K mean algorithm. As the data was already loaded I reused the seeds variable referencing the csv file. Then I plotted the ‘area’ and ‘groove length column data using a scatter plot. I then copy the objects data through ‘x = data.copy()’. This is where clustering takes place, through the variable ‘kmeans’ with an argument of 2. This is so we can make two clusters. By making a copy of x, we can use it to store clusters. Once again, we plot the clustered data with our two parameters being ‘asymmetry’ and ‘groove length’. We then use a sklearn function ‘preprocessing’, which is to standardise the datasets. The elbow method is then used to determine how many clusters we actually need, despite starting off with two clusters. Which is shown in the graph where the dip begins. In this case it being 3. Then the new cluster scatter plot is created.

# Social network analysis
#### Task 3 consists of importing the networkx library to produce a visualisation of the network and discussing the outputs. Statistics for the network that have also been calculated are also explained. 
### Task 3

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
network = pd.read_csv('social-network.csv')
```

```python
with open('social-network.csv', 'r') as nodecsv:                       
    nodereader = csv.reader(nodecsv) 
    nodes = [n for n in nodereader][1:]
```

```python
node_names = [n[0] for n in nodes]
```

```python
with open('social-network.csv', 'r') as edgecsv:
    edgereader = csv.reader(edgecsv)     
    edges = [tuple(e) for e in edgereader][1:]
```

```python
print(len(node_names)) #returns number of of objects in this case, the number of names
```

```python
print(len(edges)) # same as before, but for edges this time
```

```python
G = nx.Graph() #Empty graph
```

```python
G.add_nodes_from(node_names) # add list of nodes
G.add_edges_from(edges) # add list of edges
```

```python
print(nx.info(G)) # basic information of the graph
```

```python
print(nx.density(G)) # edge density = number of edges / maximal number of edge. This is used to describe the portion of potential connections in a network that are actual connectinos, i.e. it tells you if your network has good connections, and on a scale from 0 to 1, it's not ideally good.
```

```python
print(nx.is_weighted(G)) # Does not have weighted edges because it returned false.
```

```python
print(nx.is_negatively_weighted(G)) # There are no negative weights
```

```python
print(nx.is_connected(G)) # If this is true it means there is only one component. Otherwise, there's more than one component.
```

```python
components = nx.connected_components(G) # Gets lists of components.
largest_component = max(components, key=len) # Finds the largest component in the list.
```

```python
subgraph = G.subgraph(largest_component) #creating a subgraph of the largest component
diameter = nx.diameter(subgraph) # finding the diameter of this large component
print("Network diameter of largest component:", diameter) 
```

```python
nx.average_degree_connectivity(G) # This is the average nearest degree of nodes.
```

```python
nx.average_clustering(G) # Chooses a node at random, chooses two neighbours at random, checks if they are connected then displays mean of the local clusterings based on the average clustering coefficient.
```

```python
nx.average_neighbor_degree(G) # Returns the average degree of the neighbourhood of each node.
```

```python
nx.draw(G, node_color = 'green', node_size = 3, width = 0.25) #Visual representation of the data.
plt.show
```
